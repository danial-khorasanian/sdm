# Plot: Agents Training Performance
This plot visualizes the **Total Rewards** accumulated by three agents (**REINFORCE**, **A2C**, and **PPO**) over the course of **Time Steps**. 

We use Time Step Return instead of Episode Return because the number of steps in an episode can vary, which would misalign performance across different agents, preventing fair comparison. The Time Step Return is only updated at the end of a completed episode; for all time steps in the middle of a currently running episode, it remains constant and equal to the total return of the latest completed episode.

* **Central Line:** Represents the **mean** step returns across multiple training runs.
* **Shaded Region (Mean $\pm$ SD):** Represents the **standard deviation** of the step returns across multiple training runs. A narrower shaded region indicates more stable training process.

The raw training data is processed using two steps to generate a smoother, more interpretable curve:
1.  **Block Averaging (Downsampling):** Rewards are averaged over blocks of 200 steps to reduce noise and the total number of data points.
2.  **Moving Average (Smoothing):** A rolling mean (window of 100 steps) is applied to further smooth the curves.
